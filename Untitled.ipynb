```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#converting it into ['Action','Adventure','Fantasy','Science Fiction']
#it is a string of list that can be converted into lists using ast.literal_eval and then the function can be run 
import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]

```

```python
#we will name of top 3 actors in the film 
def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i["name"])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
#we have to exract the directory where job description is director
def director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            l.append(i["name"])
    return l
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head()
```

```python
#converting the overview which is a string to list so that it wasy to concatenate with other lists while creating the tag 
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#remove the spaces from the genres,keywords,casts and crew 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#create tag with genre,overview,cast,crew,keyword
movies['tag'] =movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
#Vectorization
 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
# Steming
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
        
         
```

```python
#gives similarity of 1st movie will all the other movie 
similarity[0]
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse = True ,key=lambda x:x[1])[1:6]
```

```python
def rec(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[index]
    movies_list = sorted(list(enumerate(distances)),reverse = True ,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```
